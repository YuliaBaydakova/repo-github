# Задача 1

 Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.


```python
#решение
a = 200
b = 800
Sr = (a+b)/2
D = ((b-a)**2)/12
print(f'Среднее значение функции равно - {Sr}, дисперсия - {D}')
```

# Задача 4

Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:
а). больше 182 см
б). больше 190 см
в). от 166 см до 190 см
г). от 166 см до 182 см
д). от 158 см до 190 см
е). не выше 150 см или не ниже 190 см
ё). не выше 150 см или не ниже 198 см
ж). ниже 166 см.


Так как для нормального распредления:
На отрезке от  −σ  до  +σ  расположено около 68 % наблюдений,
от  −2σ  до  +2σ  — 95.4 %,
и от  −3σ  до  +3σ  — 99.72 %.
По условию задачи σ = 8 см, для решения задачи воспользуемся таблицами для нормального распределения.

```python
import numpy as np
```

```python
#более 182 см, следовательно z = 1, значение из таблицы нормального распределения - 0.8413, тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост более 182 см - {round(((1- 0.8413)*100),2)}%.')
#более 190 см, следовательно z = (190-174)/8 = 2 и f(z) = 0.9772, тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост более 192 см - {round(((1-0.9772)*100),2)}%.')
#от 166 см до 190 см , следовательно z1 = (166-174)/8 = - 1 и f(z1) = 0.1587, z2 = 2 и f(z2) = 0.9772 тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 190 см - {round(((0.9772-0.1587)*100),2)}%.')
# от 166 см до 182 см , следовательно z1 = (166-174)/8 = - 1 и f(z1) = 0.1587, z2 = 1 и f(z2) = 0.8413 тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 166 см до 182 см - {round(((0.8413-0.1587)*100),2)}%.')
# от 158 см до 190 см, следовательно z1 = (158-174)/8 = - 2 и f(z1) = 0.0228, z2 = 2 и f(z2) = 0.9772 тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост от 158 см до 190 см - {round(((0.9772-0.0228)*100),2)}%.')
# не выше 150 см или не ниже 190 см, следовательно z1 = (150-174)/8 = - 3 и f(z1) = 0.013, z2 = 2 и f(z2) = 0.9772 тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 190 см - {round(((1-0.9772+0.013)*100),2)}%.')
# не выше 150 см или не ниже 198 см, следовательно z1 = (150-174)/8 = - 3 и f(z1) = 0.013, z2 = (198-174)/8 = 3 и f(z2) = 0.9987 тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост не выше 150 см или не ниже 198 см - {round(((1-0.9987+0.013)*100),2)}%.')
#ниже 166 см, следовательно z = (166-174)/8 = -1, значение из таблицы нормального распределения - 0.1587, тогда:
print (f'Вероятность того, что случайным образом выбранный взрослый человек имеет рост ниже 166 см - {round(((0.1587)*100),2)}%.')
```

# Задача 5

На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?


```python
z = 190 - 178
```

```python
#сигма выборки
sigma = np.sqrt(25)
```

```python
#искомое значение
sigma_X = z/sigma
print (f'Pост человека, равный 190 см отклоняется на {sigma_X} сигмы.')
```
